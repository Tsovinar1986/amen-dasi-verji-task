```python
import numpy as np
import matplotlib.pyplot as plt
a=np.linspace(1,2)
b=np.linspace(1,2)
def f(a,b,x):
    return (x**b +a**b)/(x**b)  
x =np.arange(-20,20,1)
plt.plot(f(1,1,x), label="a=1 b=1", color ='r')
plt.plot(f(1,2,x), label ="a=1 b=2",color = 'b')
plt.plot(f(2,1,x), label = 'a=2 b=1',color='g')
plt.xlabel('X')
plt.ylabel('F(X)')
plt.title("The graph of f(x)=${(x^b + a^b)/x^b}$ of different a and b")
plt.legend(fontsize = '10')
plt.show(f)
```

```python
import numpy as np
import matplotlib.pyplot as plt
a=np.linspace(1,1)
b=np.linspace(1,3)
def f(a,b,x):
    return (x**b +a**b)/(x**b) 
def g(a,b,x):
    return -((x**b+a**b)/(x**b))
x =np.arange(-20,20,0.3)
plt.plot(f(1,0.5,x), label="a=1 b=0.5", color ='r')
plt.plot(f(1,-0.5,x), label ="a=1 b=-0.5",color = 'b')
plt.plot(f(1,-1.5,x), label = "a=1 b=-1.5",color='g')
plt.plot(g(1,0.5,x),  color ='y')
plt.plot(g(1,-0.5,x),color = 'm')
plt.plot(g(1,-1.5,x),color='purple')
plt.xlabel('X')
plt.ylabel('F(X)')
plt.title("The graph of f(x)=${(x^b + a^b)/x^b}$ of different a and b")
plt.legend(fontsize = '10')
plt.show(f)
plt.show(g)
```
